The dataset also supports several advanced extensions:

```python
# ================================
# (D) Exploratory Data Analysis
# Amazon Electronics Ratings
# ================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
data_path = "/content/drive/MyDrive/TAMU/CSCE676/data/ratings_Electronics (1).csv.zip"

df = pd.read_csv(
    data_path,
    compression="zip",
    header=None,
    names=["userId", "productId", "rating", "timestamp"]
)

# Convert timestamp to datetime
df["datetime"] = pd.to_datetime(df["timestamp"], unit="s")

# -------------------------------
# 1. Basic Dataset Overview
# -------------------------------
n_users = df["userId"].nunique()
n_products = df["productId"].nunique()
n_ratings = len(df)

print("===== BASIC STATS =====")
print(f"Total ratings: {n_ratings}")
print(f"Unique users: {n_users}")
print(f"Unique products: {n_products}")

# -------------------------------
# 2. Rating Distribution
# -------------------------------
plt.figure()
df["rating"].value_counts().sort_index().plot(kind="bar")
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

# -------------------------------
# 3. Basket Size Distribution (reviews per user)
# -------------------------------
user_basket_sizes = df.groupby("userId").size()

plt.figure()
user_basket_sizes.hist(bins=50)
plt.yscale("log")
plt.title("Basket Size Distribution (Reviews per User)")
plt.xlabel("Number of Reviews")
plt.ylabel("Frequency (log scale)")
plt.show()

# -------------------------------
# 4. Top Product Frequency
# -------------------------------
top_products = df["productId"].value_counts().head(10)
print("\n===== TOP 10 MOST REVIEWED PRODUCTS =====")
print(top_products)

# -------------------------------
# 5. Sparsity of User-Product Matrix
# -------------------------------
sparsity = 1 - (n_ratings / (n_users * n_products))
print(f"\nUser-Product Matrix Sparsity: {sparsity:.6f}")

# -------------------------------
# 6. Temporal Activity Analysis
# -------------------------------
df["year"] = df["datetime"].dt.year

plt.figure()
df["year"].value_counts().sort_index().plot()
plt.title("Number of Ratings Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Ratings")
plt.show()

# -------------------------------
# 7. Temporal Gaps Between User Transactions
# -------------------------------
df_sorted = df.sort_values(["userId", "datetime"])
df_sorted["time_gap_days"] = df_sorted.groupby("userId")["datetime"].diff().dt.days

plt.figure()
df_sorted["time_gap_days"].dropna().clip(upper=365).hist(bins=50)
plt.title("Temporal Gaps Between User Transactions (Days, capped at 1 year)")
plt.xlabel("Days Between Reviews")
plt.ylabel("Frequency")
plt.show()
```

# (D) Exploratory Data Analysis

This section analyzes the structural and behavioral properties of the Amazon Electronics ratings dataset to inform methodological choices for association rule mining and recommendation modeling.

---

## 1. Dataset Scale and Structure

- Total Ratings: 7,824,482  
- Unique Users: 4,201,696  
- Unique Products: 476,002  
- User–Product Matrix Sparsity: 0.999996  

The interaction matrix is extremely sparse, meaning that only a tiny fraction of possible user–product pairs contain interactions. This confirms that traditional dense modeling approaches are inappropriate and motivates sparse-aware algorithms and graph-based methods.

---

## 2. Rating Distribution

The rating distribution is highly skewed toward 4 and 5 stars, with 5-star ratings dominating the dataset.

Implications:
- Strong positivity bias.
- Reduced variance in rating prediction.
- Binary interaction modeling (e.g., rating ≥ 4) may be more appropriate than regression.
- Association rules may be influenced by high-rating dominance.

---

## 3. Basket Size Distribution (User Activity)

The number of reviews per user follows an extreme long-tail distribution:

- The majority of users review only 1–5 products.
- A very small number of users review hundreds of products.
- The distribution spans up to ~500+ reviews for rare power users.

Implications:
- Transaction baskets are typically very small.
- Frequent itemset mining must account for sparse co-occurrence.
- Filtering users with minimum interaction thresholds is necessary.
- Power users may disproportionately influence rule discovery.

---

## 4. Product Popularity Distribution

The top products receive between ~9,000 and ~18,000 reviews, while most products receive very few.

Implications:
- Strong popularity bias.
- High-support itemsets will likely involve only dominant products.
- Support thresholds must be tuned carefully to avoid trivial rules.

---

## 5. Temporal Growth Patterns

The number of ratings increases sharply over time, particularly after 2008, with a major peak around 2012–2013.

Implications:
- Platform adoption accelerated during this period.
- User behavior may differ across time.
- Temporal segmentation may reveal evolving co-purchase patterns.

---

## 6. Temporal Gaps Between User Transactions

Temporal gaps between reviews show:

- A large concentration of short intervals (rapid repeated activity).
- A broad spread across weeks and months.
- A visible spike near 365 days due to the imposed cap.

Implications:
- Many users exhibit bursty behavior.
- Sequential modeling may capture time-dependent purchasing behavior.
- Static association rules ignore temporal ordering.

---

## Summary of Structural Findings

The dataset exhibits:

- Extreme sparsity
- Strong long-tail distributions
- Popularity concentration
- Rating inflation
- Temporal growth and bursty activity

These characteristics directly influence algorithm selection and evaluation strategies in subsequent phases.

# (E) Initial Insights and Direction

## Key Observations

1. Extreme sparsity (0.999996) makes dense collaborative filtering unsuitable.
2. Most users interact with very few products.
3. Most products appear in only a small fraction of user baskets.
4. Rating inflation reduces the effectiveness of regression-based prediction.
5. Popular products dominate high-frequency counts.
6. User activity shows bursty temporal patterns.

---

## Hypotheses

### Hypothesis 1: Support Sensitivity Hypothesis
High minimum support thresholds will primarily identify rules involving highly popular products, while lower thresholds may uncover niche but meaningful co-purchase relationships.

### Hypothesis 2: Popularity Bias Hypothesis
Association rules derived without popularity normalization will over-represent dominant products, limiting personalization quality.

### Hypothesis 3: Interpretability vs Predictive Trade-Off
Association rule-based recommendations will provide interpretable co-occurrence explanations but will underperform embedding-based recommendation models in predictive accuracy.

### Hypothesis 4: Hybrid Enhancement Hypothesis
Incorporating rule-derived co-occurrence signals into collaborative filtering models will improve performance relative to pure matrix factorization.

---

## Research Questions (RQs)

### RQ1 — Support Threshold Analysis
How does varying minimum support affect:
- Number of discovered rules?
- Average lift?
- Rule diversity?
- Long-tail product coverage?

### RQ2 — Popularity Bias Assessment
Do high-confidence rules primarily involve top-1% most popular products?

### RQ3 — Rule-Based Recommendation Performance
How does a rule-based recommendation system compare to matrix factorization in:
- Precision@K
- Recall@K
- Coverage of long-tail products?

### RQ4 — Hybrid Modeling
Does integrating association rule confidence scores as additional features improve recommendation performance?
